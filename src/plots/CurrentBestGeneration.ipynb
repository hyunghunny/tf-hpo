```python
def get_current_best(test_acuracies):
    count = len(test_acuracies)
    current_best = 0
    best_acc = []
    for i in range(count):
        test_acc = test_acuracies[i]
        if current_best < test_acc:
            current_best = test_acc

        best_acc.append(current_best)
    return best_acc
```
